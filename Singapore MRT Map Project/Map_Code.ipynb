```python
import pandas as pd
df = pd.read_csv('mrtsg.csv')
df.head()
```

```python
df = df.drop(['OBJECTID','STN_NAME','X','Y'], axis=1)
df.head()
```

```python
df = df[df.COLOR != 'OTHERS']
```

```python
#Co-ordinates of Singapore
latitude = 1.3521
longitude = 103.8198
```

```python
import folium
```

```python
singapore_map = folium.Map(location=[latitude, longitude], zoom_start=12)

for lat, lng, label, c in zip(df.Latitude, df.Longitude, df.STN_NO, df.COLOR):
    folium.CircleMarker(
        [lat, lng],
        radius=5, 
        color='black',
        fill=True,
        popup=label,
        fill_color=str(c),
        fill_opacity=1
    ).add_to(singapore_map)

singapore_map
```
